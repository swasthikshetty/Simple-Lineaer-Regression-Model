```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn import linear_model

data=pd.read_csv("C:/Users/Swathi/Desktop/Excelr Classes/Assignments/3.Simplelinear regression/Salary_Data.csv")
data.head()
```

```python
%matplotlib inline
plt.scatter(data['YearsExperience'],data['Salary'],color='red',marker="o")
plt.title("Data Set")
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.show()
```

```python

slr_model=smf.ols("data['Salary']~data['YearsExperience']",data=data).fit()
print(slr_model.summary())

predict=slr_model.predict(data.iloc[:,0])

plt.scatter(data['YearsExperience'],data['Salary'],color='red',marker="o")
plt.plot(data['YearsExperience'],predict,color='blue')
plt.title("SLR Data set")
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.show()
```

```python
# Log Model

log_model=smf.ols("data['Salary']~np.log(data['YearsExperience'])", data=data).fit()
print(log_model.summary())

log_predict=log_model.predict(pd.DataFrame(data['YearsExperience']))

plt.scatter(data['YearsExperience'],data['Salary'],color='green',marker="o")
plt.plot(data['YearsExperience'],log_predict,color='black')
plt.title("log Data model")
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.show()
```

```python
#Exponential Model


exp_model=smf.ols("np.log(data['Salary'])~data['YearsExperience']", data=data).fit()
print(exp_model.summary())

exp_predict=exp_model.predict(pd.DataFrame(data['YearsExperience']))

pred_exp=np.exp(exp_predict)  # as we have used log(Salary) in preparing model so we need to convert it back


plt.scatter(data['YearsExperience'],data['Salary'],color='black',marker="*")
plt.plot(data['YearsExperience'],np.exp(exp_predict),color='yellow')
plt.title("Exponential_model plotting")
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.show()
```

```python
slr_regression=linear_model.LinearRegression()
slr_regression.fit(data[['YearsExperience']],data.Salary)


flag=True
while(flag):
    myinput=float(input("Enter Years of Experience"))
    myoutput=slr_regression.predict([[myinput]])
    print("Salary Hike precited using SLR is: ",myoutput)
    flag=int(input("press 1 to continue or press 0 to exit: "))
```
